```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
res = []
with open('learning_log.txt') as file:
    for line in file:
        res.append(line.rstrip())
```

```python
res[0], res[1]
```

```python

float(res[0].split(':')[1])
```

```python
res[1].split(':')[1][2:-1].split(',')
```

```python
fl, l1offset, l1size, loss = [], [], [], []

for i, line in enumerate(res):
    if len(line.split(':')) > 1:
        if i % 2 == 0:
            loss.append(float(line.split(':')[1]))
        else:
            losses = line.split(':')[1][2:-1].split(',')
            fl.append(float(losses[0]))
            l1offset.append(float(losses[1]))
            l1size.append(float(losses[2]))
            
```

```python
plt.figure(figsize=(10, 10))
plt.plot(range(len(loss)), np.log(loss))
plt.xlabel('Train step')
plt.ylabel('log Loss')
plt.title('Overall loss')
plt.grid()
```

```python
plt.figure(figsize=(10, 10))
plt.plot(range(len(l1offset)), l1offset)
plt.xlabel('Train step')
plt.ylabel('Loss')
plt.title('Offset loss')
plt.grid()
```

```python
plt.figure(figsize=(10, 10))
plt.plot(range(len(l1size)), l1size)
plt.xlabel('Train step')
plt.ylabel('log Loss')
plt.title('Size loss')
plt.grid()
```

```python
plt.figure(figsize=(10, 10))
plt.plot(range(len(fl)), np.log(fl))
plt.xlabel('Train step')
plt.ylabel('log Loss')
plt.title('Heatmap loss')
plt.grid()
```
